```python
from pathlib import Path
import zipfile
from google.colab import files

project_dir = Path('/content/ecommerce_data_pipeline')

def safe_extract(zip_path, dest):
    dest = Path(dest).resolve()
    with zipfile.ZipFile(zip_path, 'r') as zipf:
        for member in zipf.infolist():
            target = (dest / member.filename).resolve()
            try:
                target.relative_to(dest)
            except ValueError:
                raise RuntimeError(f'ZIP 内存在不安全路径: {member.filename}')
        zipf.extractall(dest)

if not project_dir.exists():
    print('未找到 /content/ecommerce_data_pipeline，请上传完整项目 ZIP。')
    uploaded = files.upload()
    if not uploaded:
        raise RuntimeError('未上传项目 ZIP。')
    zip_name = next(iter(uploaded.keys()))
    safe_extract(zip_name, '/content')
    if not project_dir.exists():
        candidates = [p for p in Path('/content').glob('*/colab_runner.py')]
        if candidates:
            candidates[0].parent.rename(project_dir)
    if not project_dir.exists():
        raise RuntimeError('项目 ZIP 解压后未找到 ecommerce_data_pipeline/colab_runner.py。')
else:
    print('已找到项目目录:', project_dir)

%cd /content/ecommerce_data_pipeline
```
